```python
import pandas as pd
```

# Census Data

```python
df = pd.read_csv("./census2021/98-401-X2021003CI_English_CSV_data.csv", encoding='ISO-8859-1')
```

```python
df.head()
```

```python
print(df.columns.values)
```

```python
print(sorted(df['GEO_NAME'].unique()))
```

```python
income2019_df = df[df['CHARACTERISTIC_NAME'] == '    Average total income in 2019 among recipients ($)']
income2019_df.head()
```

```python
income2020_df = df[df['CHARACTERISTIC_NAME'] == '    Average total income in 2020 among recipients ($)']
income2020_df.head()
```

```python
income2019_df.info()
```

```python
income1920_df = pd.merge(income2019_df[['GEO_NAME', 'C14_RATE_LOW_CI_MEN+', 'C15_RATE_LOW_CI_WOMEN+']], 
                     income2020_df[['GEO_NAME', 'C14_RATE_LOW_CI_MEN+', 'C15_RATE_LOW_CI_WOMEN+']], 
                     on='GEO_NAME', 
                     suffixes=('_19', '_20'))
```

```python
income1920_df.info()
```

```python
income1920_df['C14_RATE_LOW_CI_MEN+_change'] = ((income1920_df['C14_RATE_LOW_CI_MEN+_20'] - income1920_df['C14_RATE_LOW_CI_MEN+_19']) / income1920_df['C14_RATE_LOW_CI_MEN+_19']) * 100
```

```python
income1920_df['C15_RATE_LOW_CI_WOMEN+_change'] = ((income1920_df['C15_RATE_LOW_CI_WOMEN+_20'] - income1920_df['C15_RATE_LOW_CI_WOMEN+_19']) / income1920_df['C15_RATE_LOW_CI_WOMEN+_19']) * 100
```

```python
income_df = income1920_df[['GEO_NAME', 'C14_RATE_LOW_CI_MEN+_change', 'C15_RATE_LOW_CI_WOMEN+_change', 'C14_RATE_LOW_CI_MEN+_19', 'C14_RATE_LOW_CI_MEN+_20', 'C15_RATE_LOW_CI_WOMEN+_19', 'C15_RATE_LOW_CI_WOMEN+_20']]
```

```python
income_df.rename(columns={'GEO_NAME': 'geo_name', 'C14_RATE_LOW_CI_MEN+_change': 'income_change_men', 'C15_RATE_LOW_CI_WOMEN+_change': 'income_change_women', 'C14_RATE_LOW_CI_MEN+_19': 'income_men_2019', 'C14_RATE_LOW_CI_MEN+_20': 'income_men_2020', 'C15_RATE_LOW_CI_WOMEN+_19': 'income_women_2019', 'C15_RATE_LOW_CI_WOMEN+_20': 'income_women_2020'})
```

```python
income_df.to_csv('./census2021/income.csv', index=False)
```
